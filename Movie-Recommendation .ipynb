```python
z=cv.fit_transform(y['tags'].values.astype('U')).toarray()
from sklearn.metrics.pairwise import cosine_similarity
cs=cosine_similarity(z)
```

```python
def recommend(movies):
    index=y[y['title']==movies].index[0]
    sim=sorted(list(enumerate(cs[index])),reverse=True,key=lambda z:z[1])
    for i in sim[1:6]:
        print(y.iloc[i[0]].title)
```
